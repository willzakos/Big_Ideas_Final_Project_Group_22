One of the most important factors of any neighborhood is safety. Safety can be defined in this context as having a low rate of crime within the neighborhood. Using data from the WPRDC, Brookline is shown to have a consistently low rate of crime in comparison to other neighborhoods within the city of Pittsburgh.

```python
import pandas as pd
import numpy as np
file = "crime_stats.csv"
df = pd.read_csv(file, low_memory=False)
df2 = pd.read_csv("popStats.csv")
df2 = df2.iloc[:,[0,1]]
df2
```

First, I input two datasets from the WPRDC into python, along with numpy and pandas. I then take the population data by neighborhood and turn that into its own dataframe. Then, I strip the amount of police incidents per neighborhood from the Crime Statistics dataset. I then merge these two datasets together resulting in a dataframe that contains a list of all neighborhoods. The end result is a dataframe that contains every Neighborhood along with its population and the amount of crimes that were reported within the last month.

```python
nb_col = df.Neighborhood



nb_col_val_counts = nb_col.value_counts()

df_new = nb_col_val_counts.to_frame()

df_new = pd.merge(df_new, df2[["Neighborhood","totalPop"]], on="Neighborhood", how="right")
df_new
```

This graph below demonstrates the amount of times each neighborhood is mentioned in the crime reports database (count) along with the total population of each neighborhood (totalPop). This graph clearly demonstrates the safety of Brookline, as despite having an incredibly large population it has a comparitively low crime rate.

```python
df_new.plot(kind='bar', stacked=True, x="Neighborhood", title="Amount of crime reports made in each Pittsburgh neighborhood", figsize=(15, 10))

```

This graph contains the findings of all neighborhoods within the city of Pittsburgh, along with their population size and crimes reported within the past month. To focus in further on Brookline, the bottom graph has been trimmed to only contain other neighborhoods with a similar population size (Above 12,000 residents)

```python
df_trim = df_new.query('totalPop > 12000')
df_trim = df_trim.sort_values(by='count')
df_trim.plot(kind='bar', stacked=True, x="Neighborhood", title="Amount of crime reports made in each Pittsburgh neighborhood with a population greater than 12000", figsize=(15, 10))
```

The graph above shows the results with neighborhoods with a population greater than 12,000 residents. As you can see, Brookline has the lowest crime rate out of all Neighborhoods that contain a similar population size (Above 12,000 residents). This shows Brooklines safety and security as a neighborhood, specifically compared to other large neighborhoods of a similar size.
